```python
import os
import pandas as pd

directory_path = '/home/randolpwanjiru/DSP4380/Tabular/Tabular_new/csv_files/train'
file_paths = [os.path.join(directory_path, filename) for filename in os.listdir(directory_path)]
chunk_size = 5000
chunks_to_read_per_file = 3  # Adjust the number of chunks to read

# Initialize an empty list to store DataFrames
all_dataframes = []

# Initialize an empty list to store file names
file_names = []

# Process files in chunks
for file_path in file_paths:
    print(f'Processing file: {file_path}')
    chunk_iter = pd.read_csv(file_path, chunksize=chunk_size, low_memory=False)
    for i, chunk in enumerate(chunk_iter):
        if i >= chunks_to_read_per_file:
            break  # Exit the loop if the desired number of chunks have been read
        
        # Data cleaning and processing...
        chunk.columns = [col.lower() for col in chunk.columns]
        
        # Display sample of rows from the processed chunk
        print(f'Sample of rows from Chunk {i}:')
        print(chunk.sample(n=5))  # Adjust n to the number of rows
        
        # Append the chunk DataFrame to the list
        all_dataframes.append(chunk)
        
        # Append the file name to the list for each row in the chunk
        file_names.extend([os.path.basename(file_path)] * len(chunk))

# Concatenate all DataFrames into a single DataFrame
df = pd.concat(all_dataframes)

# Create a DataFrame for file_names and reset the index
file_names_df = pd.DataFrame({'source_file': file_names}).reset_index(drop=True)

# Reset the index of df
df.reset_index(drop=True, inplace=True)

# Join the source_file column to df
df = pd.concat([df, file_names_df], axis=1)
```

```python
# Generate descriptive statistics for the DataFrame
description = df.describe()

# Display the descriptive statistics
print(description)
```

```python
# Check for missing values in each column
missing_values = df.isna().sum()

# Display columns with missing values
columns_with_missing_values = missing_values[missing_values > 0]
print("Columns with missing values:")
print(columns_with_missing_values)

# Total number of missing values in the DataFrame
total_missing_values = missing_values.sum()
print("\nTotal number of missing values:", total_missing_values)
#number of rows and columns in the DataFrame
num_rows, num_columns = df.shape


print("Number of rows:", num_rows)
print("Number of columns:", num_columns)

# Check for duplicate records
duplicate_records = df.duplicated()

# Count the number of duplicate records
num_duplicates = duplicate_records.sum()


print("Number of duplicate records:", num_duplicates)

```

```python
# Select only numeric columns
numeric_columns = df.select_dtypes(include='number')

# Impute missing values in numeric columns with mean
cleaned_df = df.copy()
cleaned_df[numeric_columns.columns] = numeric_columns.fillna(numeric_columns.mean())


# Check if missing values were successfully imputed
imputation_check = (df[numeric_columns.columns].isna() == cleaned_df[numeric_columns.columns].isna()).all().all()

# Print the result
if imputation_check:
    print("Missing values were successfully imputed with mean in numeric columns.")
else:
    print("There was a problem with imputing missing values with mean in numeric columns.")
```
