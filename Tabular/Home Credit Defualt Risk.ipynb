```python
new_file_paths = [
    'combined_data_chunk_0.csv',
    'combined_data_chunk_1.csv',
    # Add more file paths if necessary
]

# List to store DataFrames
new_dfs = []

# Read new CSV files and append its DataFrame to list
for new_file_path in new_file_paths:
    new_df = pd.read_csv(new_file_path)
    new_dfs.append(new_df)
```

```python
# Display each DataFrame in the list
for i, df in enumerate(new_dfs):
    print(f"DataFrame {i+1}:")
    display(df)  # This will display the DataFrame nicely formatted in a Jupyter Notebook
```

```python
second_df = new_dfs[1]

# Check for NaN values in the DataFrame
nan_counts = second_df.isna().sum()  #this dataset needs to be cleaned 
print("Number of NaN values in each column:\n")
print(nan_counts)

first_df = new_dfs[0] #this data set is fine no cleaning necessary
nan_counts = first_df.isna().sum()
print("Number of NaN values in each column:\n")
print(nan_counts)


```
